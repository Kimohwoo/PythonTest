```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
```

```python
from matplotlib import rc, font_manager
fontname = 'c:/windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=fontname).get_name()
rc('font', family=font_name)

plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.figsize']=(12,9)
```

```python
tips = sns.load_dataset('tips')
tips.head()
# tips.tail()
```

# catplot
- 범주를 나누어서 그려줌
- 참조사이트 : https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', hue='smoker', split=True)
```

# Box plot

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar', hue='smoker')
```

```python
sns.catplot(x='day', data=tips, kind='count', hue='sex')
```

```python
titanic = sns.load_dataset('titanic')
sns.catplot(x='class', y='survived', data=titanic, col='sex', kind='bar')
```

#pairplot
- 그리드 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 표시
- 숫자형 column에 대해서만 그려줌

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='size')
```

# lmplot
- lmplot: 컬럼간의 선형관계를 확인할 수 있다
- outlier 짐작가능

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, col='day', col_wrap=2)
```

# joinplot
- scatter(산점도)와 histogram(분포)을 동시에 그려주며 숫자형 데이터만 표현 가능

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```
