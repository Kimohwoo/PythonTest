```python
import pandas as pd
import numpy as np
from folium import plugins
df = pd.read_csv('data/CCTV_20190917.csv')
df1 = df.loc[df['카메라대수']>0,['소재지도로명주소', '위도', '경도']]

popups = df1['소재지도로명주소'].to_list()
log = df1['경도'].to_list()
lat = df1['위도'].to_list()
lat

data = np.array([lat, log]).T

m = folium.Map(location=[35.1803305, 129.0516257], zoom_start=17)
plugins.MarkerCluster(data, popups=popups).add_to(m)
m
```

```python
import requests
import json

r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_grep = json.loads(c)
seoul_grep
```

```python
m = folium.Map(location=[37.557222485451305, 127.11879551821994], zoom_start=11)
folium.GeoJson(seoul_grep, name='지역구').add_to(m)
m
```

```python
cb = pd.read_csv('data/CoffeeBean.csv')
add = pd.DataFrame(cb.address.apply(lambda v : v.split()[:2]).to_list(), columns=['시도', '구군'])
# add.시도.unique()
add
```

```python
addr_aliase = {'서울':'서울특별시', '서울시':'서울특별시'}
addr_aliase
add['시도']=add['시도'].apply(lambda v: addr_aliase.get(v, v))
add.시도.unique()
```

```python
addr2 = add.loc[add['시도']=='서울특별시', :]
addr2
```

```python
addr_group = addr2.groupby('구군')['시도'].count()
addr_group
```

```python
m.choropleth(geo_data=seoul_grep, data=addr_group,fill_color="YlOrRd", fill_opacity=0.7, line_opacity=0.2, key_on='properties.name',  legend_name='지역구별 커피 업종수')
m
```
