# heatmap
- 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지 위에 열분포 형태의 비쥬얼한 그래픽 출력기능

```python
data = np.random.rand(5,6)
data
```

```python
sns.heatmap(data, annot=True)
```

```python
pivot = tips.pivot_table(index='day', columns='size', values='tip')
pivot
```

```python
sns.heatmap(pivot, cmap='Greens', annot=True)
```

```python
tips_corr = tips.corr()
```

```python
sns.heatmap(tips_corr, annot=True, cmap='YlGnBu')
```
